## I(E)

```python
E = np.arange(0, 1, 0.001)
I = arcs.model.I_E(E)
plt.figure()
plt.title("I(E)")
plt.xlabel('Energy (eV)')
plt.plot(E,I)
```

## I(t) - Ikeda-Carpenter function

```python
E = 100 #meV
```

```python
I_t = arcs.model.IC_atE(E/1e3)
```

```python
t = np.arange(0,200, 0.1)
I = I_t(t)
plt.figure()
plt.title("I(t)")
plt.xlabel('t (microsecond)')
plt.plot(t,I)
```

```python
plt.figure()
plt.title("I(t)")
plt.xlabel('t (microsecond)')
plt.semilogy(t,I)
```

## I(t) through Fermi chopper

```python
t = np.arange(0, 200., .1)
t0 = 10.
from fermichopper import FermiChopper as base
class FC(base):
    frequency = 600 #Hz
    slit_thickness = 1.5 #mm
    radius = 50 #mm
I = arcs.model.time_distrib_thru_fc(t, t0, E/1e3, FC())
plt.figure()
plt.title("I(t)")
plt.xlabel('t (microsecond)')
plt.plot(t,I)
```

```python
plt.figure()
plt.title("I(t)")
plt.xlabel('t (microsecond)')
plt.semilogy(t,I)
```
